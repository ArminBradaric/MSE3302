# **<u>Abstract</u>**

<p align = "justify">$\;\;\;\;\;\;$Fluoroscopy machines provide real time medical imaging of a patient’s interior to allow for surgeons to preform less invasive surgical operations. The developed miniature fluoroscopy device emulates the true scale machine to allow for hands on training for medical practitioners. With this device medical practitioners can more easily and cost efficiently be trained on floroscopy machines.  It is controlled through an Arduino microcontroller utilizing a joystick to translate the bed and rotate the C-Arm. Accelerometers are used to track the location of the bed, and C-Arm which allong with software can mimic the view a larger floroscopy machine would have of a live patient. The device is cheap and easy to manufacture with the vast majority of its parts being 3D printed. The use of an Arduino uno alongside a MX1508 motor controller provide cheap and readily available electronics to facilitate the rapid implementation of this device to medical practitioners around the world.


# <u>**Introduction:**</u>


<p align = "justify">$\;\;\;\;\;\;$Fluoroscopy is a medical procedure that makes a real-time video of the movements inside a part of the body by passing x-rays through the body over a period. This device is composed a moving bed and rotating C-arm, the end C-arm contain an x-ray emitter and detector. A continuous X-ray is sent through the body part being examined while the C-arm of the fluoroscopy device is rotated and translated to capture different perspectives of the body. The images are transmitted to a monitor in real time for analysis. Fluoroscopy is used for many different types of examinations, such as tracking the movement of the intestines, blood flow through arteries, visualization of joints, and many more. These examinations are essential for surgeons to use when diagnosing diseases and helpful for performing surgery.

<p align = "justify">$\;\;\;\;\;\;$Fluoroscopy devices are complex and expensive to operate and expose both the doctor and patient to extensive radiation, which is why it is important to design surgical simulators. Having a simulator will allow for medical practitioners to be trained in an efficient, cost-effective manner. This report details the design and manufacturing of a low-cost fluoroscopy simulator that can be mass distributed to medical facilities, to aid in the training of certified medical practitioners.</p>






# **<u>User Control Method:</u>**



<p align = "justify">$\;\;\;\;\;\;$Selecting to use a joystick for the purpose of this prototype over other control options, such as pushbuttons, or potentiometers was a major discussion before the software was implemented. The choice to use a joystick was made after research into actual fluoroscopy C-arm devices found that most of their actuated systems are controlled with the use of joysticks. This makes sense as it allows for more conventional and convenient user control over the use of binary devices such as pushbuttons that may seem less intuitive to use in this application.  

<p align = "justify">$\;\;\;\;\;\;$The joystick requires the use of less wires than setting up 4 push buttons for the control of 2 motors. This allows for simpler identification of wiring problems should an issue arise during use.</p>

## <i>Electrical:</i>


<p align = "justify">$\;\;\;\;\;\;$The electronic components utilized within this fluoroscopy simulator consist of an Arduino Uno microcontroller, two six-volt DC geared motors, one MX1508 motor controller, one Arduino five-volt power supply module with a nine-volt battery and one PS2 Joystick game controller for Arduino.
The microcontroller used for this project allows for easy control of the systems electronics due to its simple user interface and widely available resources such as libraries and supported hardware. This microcontroller accepts both digital and analog inputs from the Joystick and outputs both digital and analog signals to provide motor control, making it a simple, cost-effective solution for this device.
The motors for this project are six-volt DC geared motors that operate at 20 RPM but can be stepped down simply by lowering its input voltage.

<p align = "justify">$\;\;\;\;\;\;$The motors used in this system were rated for six-volts and so could not be directly powered by the nine-volt battery. This introduced the requirement for a device to step down the voltage to something less than or equal to six-volts while still being able to power the Arduino. The Arduino five-volt power supply module fit this role perfectly. It could take the input voltage of nine-volts in its input barrel connector and step it down to five-volts at its 5V pins that the motors and Arduino could use.The modual's pinouts are pictured in figure 11.

## <i>Arduino Software Setup:</i>



<p align = "justify">$\;\;\;\;\;\;$Knowing that this device will most likely be used by medical students, it is important that the code is simple and easy to follow should a consumer use the prototype in the future and wish to modify the software themselves. To reduce unnecessary complications the software implemented in this project requires the use of 4 digital I/O pins to control the motor controller, as well as 2 analog input pins to read the joystick position on the Arduino Uno. The joystick’s x and y axis analog readings will control the directions of each motor respectively.

<p align = "justify">$\;\;\;\;\;\;$Describing the code in Figure 13, all 4 digital output pins are first defined at the beginning of the program and are given the names DIR_PIN1, DIR_PIN2, DIR_PIN2, DIR2_PIN2. These pins control which direction each motor will spin depending on which of the 2 pairs are set high and low. Next, the joystick x and y pins are defined from their input A0, and A1. Since it is beneficial that stick drift be eliminated through the software, dead zones are implemented as a threshold and are defined as the analog readings of 100 and 900. These represent the minimum and maximum thresholds for the joystick. Moving the stick past these limits will cause the motors to move.

<p align="justify">$\;\;\;\;\;\;$Within the setup, the digital I/O pins are set to output using the function pinMode(pinName, OUTPUT) and the analog pins are set to input in the same manner.

<p align="justify">$\;\;\;\;\;\;$Next, two separate but similar functions were made demonstrated in Figure 14 that take a direction as an argument (either 1 for forward or 0 for backward). These functions set the digital I/O pins to either high or low depending on the inputted direction. For example, in motor 1, if the direction input to the function is 1 (forward) then DIR_PIN1 - which was defined earlier - will be set to high, and its complement DIR2_PIN1 will be set to low. These motor controller commands allow for the motor to spin in the forward direction. Should 0 be inputted to the function then the pins are reversed, and the motor will spin in the opposite direction (backwards). ### **<i>Main Loop:</i>**
<p align = "justify">$\;\;\;\;\;\;$Within the main loop pictured in Figure 15, the joystick x and y values are continuously read every cycle and are assigned to the variables joystickXValue, and JoystickYValue respectively. Conditional statements are then used along with the threshold values to tell the program when to execute the ‘setMotorDirection()’ function. For example, the first if{} statement says that if the x value read from the joystick lies below the minimum threshold of 100, corresponding to the user moving the joystick all the way to the right, then the ‘motorDirectionFunction’ will be executed with a 0 input, corresponding to the “backwards directions”. Should the user move past the 900 analog value threshold, corresponding to the stick moving all the way to the right, then the motor will begin to move in the opposing direction. These conditional statements are repeated for the y direction of the other motor. To facilitate this, an else{} statement is included within these if{} statement blocks to tell the program to set the direction pins all to low, should the joystick not move past any threshold in a given clock cycle.

# **<u>Discussion:</u>**

## <i>Difficulties:</i>



<p align = "justify">$\;\;\;\;\;\;$The design and construction of this device presented several significant challenges. Foremost was the brainstorming phase, where achieving unanimous agreement while effectively dividing group work without diluting the universal vision of the final product proved demanding. Additionally, various peripheral issues arose, such as multiple tolerancing challenges impacting the printing of interference fits, introducing bearing design complexities, and difficulty in printing slots in the gear and rotor to accommodate the N20 DC motor shaft. Moreover, challenges surfaced during the implementation of accelerometers and the integration of their data with the Slicer 3D app.

<p align = "justify">$\;\;\;\;\;\;$Tolerancing issues among the three different printers utilized in the production of this prototype led to issues with interference fits due to variations in printer head tolerancing, therefore the method of splitting parts between different printers resulted in suboptimal outcomes. To address this issue, certain components, such as the two parts of the rack, were reprinted using the same printer to eliminate printer head inconsistencies. However, larger parts like the main frame and body, which were impractical to reprint entirely, required the filing down of male interference parts the application of glue to ensure secure bonding.

<p align = "justify">$\;\;\;\;\;\;$Designing around the slots for the d-shafts of the N20 DC motor posed significant challenges. The combination of low percentage infill parts, poor tolerancing, and the relatively high torque requirements of the C-Arm resulted in stress concentrations in the PLA around the motor shaft, causing deformation and free spinning. One proposed solution was to laser cut the rotor from a rigid material like acrylic to prevent deformation of the slot. However, this process was deemed too time-consuming. Instead, the rotor was reprinted with a higher infill, and the C-Arm infill was reduced to decrease the torque required for rotation.

<p align = "justify">$\;\;\;\;\;\;$Splitting up the 3D printing of parts with initially poor communication also led to inconsistencies with percentage infill usage. For example, the male and female sections of the C-Arm were initially printed on two separate printers by two separate group members. Poor communication had led to one group member rationalizing that a higher infill, approximately 40%, would be more effective as it would prevent the set screw from creating surface defects in the side of the C-Arm. However, the other group member had reasoned that a 10% infill was justified because it would save on material costs and printing time. Due to this lapse in team coordination, the initial prototype of the C-Arm was lopsided and unbalanced because of the weight difference of each section caused by the infill inconsistencies. To rectify this, the group simply decided that a 10% infill part was the better choice, and it was simply reprinted.
<p align = "justify">$\;\;\;\;\;\;$When crafting the bearing, significant difficulty arose when attempting to position the ball bearings within their housing. This issue can be attributed to an initial design error, where the gap between the base wall and the inner bearing ring was excessively large. This allowed the ball bearings to slide beneath the inner bearing wheel, resulting in an uneven bearing and substandard performance. Consequently, the entire C-Arm appeared unbalanced and improperly aligned due to this flaw.

<p align = "justify">$\;\;\;\;\;\;$In addition to the mechanical difficulties, we also ran into considerable troubles when attempting to interface the accelerometer with the Slicer 3D app. These struggles could be organized into two categories: connection and software. The connection issues included finding a Bluetooth connection between the accelerometer and our computer and launching a server correctly with the Plus Server app. The instructions provided valuable steps to correctly navigating the Bluetooth connectivity issues, while we made independent educated adjustments to the Plus Server file to launch the server correctly. This portion provided a valuable learning experience as it forced us to think critically about what we were trying to do and how to best achieve that. The software issues proved more difficult, as we were unable to overcome them. The most important of these issues was the inability of the Slicer app/Plus Server to obtain the data quick enough causing a backlog and eventual crash of the C-arm model. This made it impossible to gather meaningful data with this method.

## <i>Next Steps:</i>


<p align = "justify">$\;\;\;\;\;\;$Overall, the production of the first functional prototype was successful. However, there remains a huge array of possibilities for improvement in future iterations. Most notably, the set screw method of stalling the C-Arm translation leaves much to be desired in terms of user friendliness and is need of a far more elegant implementation than using a screwdriver. The device could also benefit from both the replacement of the current actuators (N20 DC motors) and the inclusion of another actuator that produces C-Arm translation as opposed to that being the only degree of freedom absent from electrical actuation. The Bed would also be improved with another level of vertical support that will reduce the friction is undergoes and allow for smoother translation. The method of production could also benefit from change, as mass producing a 3D-printed device as large and complex as this for a reasonable price is unrealistic. Finally, manual speed control would bring the whole device together and have it truly embody a final product.

<p align = "justify">$\;\;\;\;\;\;$To improve the method of maintaining the C-Arm Position, there are two possibilities, either maintain a simpler design like the set screw and simply make it more user-friendly or implement a transducer that takes an input from a joystick sensor and actuates a semi-circular rack and pinion device. Having an actuated method of C-Arm translation, although a more complex solution, would provide the user with a higher resolution movement than human controlled. It would also make the simulation device a more effective simulator, which of course is the final goal. It would only serve to break the suspension of disbelief of the user practicing their fluoroscopy skills to be looking at a monitor projecting the simulated x-ray imaging whilst manipulating the joystick to then have to look over at an extremely scaled down physical device and start manipulating that by hand in a fashion unlike the real fluoroscopic devices.

<p align = "justify">$\;\;\;\;\;\;$To improve the overall smoothness and functionality of the simulator, implementing high-resolution stepper motors in place of the cheap and prone to breaking N20 DC Geared motors would provide the user with far more control over the multiple degrees of freedom present in the device's functions. It would also serve to reduce choppy movement. Along with higher resolution actuators, the bed mechanism could benefit from more vertical support on the bed body, reducing friction and stress concentrations on the small section at the pinion which inevitably cause choppy movements due to uneven surfaces.

<p align = "justify">$\;\;\;\;\;\;$Speed control has not been implemented in the software as this is seen as a next step in future models. While the benefits to the effectiveness of the simulation of implementing speed control of the actuators are immense, the changes would be relatively simple. The program would only have to be modified so that a PWM signal could be sent to the motor controller rather than just a high/low to allow the motor to operate at a lower voltage. Since the Arduino components used already support this PWM, and since motor speed is proportional to input voltage, this would likely be the most effective method of creating speed control. This would best be accomplished with a potentiometer read as analog input as it would allow for easy adjustment of the PWM value.
